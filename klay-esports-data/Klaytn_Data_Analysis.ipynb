```python
def featuring(data) :
   
    #결측값(-) 제거
    data['Solo Kills'] = data['Solo Kills'].apply(lambda x:0 if x=="-" else x)
    meanavg = np.mean(data['Avg deaths'])
    data['Avg deaths'].fillna(meanavg)
    
     #scaling
    object_ = ['Player','Position']
    data_noneed = data[object_]
    data_toscale = data.drop(object_, axis=1)

    scaler = StandardScaler()
    scaler.fit(data_toscale)
    scaled_matrix = scaler.transform(data_toscale)
    scaled_data = pd.DataFrame(scaled_matrix)
    scaled_data.columns = list(data_toscale)

    data = pd.concat([data_noneed, scaled_data], axis=1)
    
    
    #중복되는 선수 제거
    unique_list = data.groupby(['Player'])['Games'].max()
    unique_df = pd.DataFrame(unique_list).reset_index()
    unique_df2 = pd.merge(data, unique_df, on=["Player","Games"],how="right")
    data = unique_df2
    return data
    
a = featuring(data)
a.head()
```

```python
def updateDealingMachine(originalData) :
    data = featuring(originalData)
    a = data['DPM'].quantile(q=0.95, interpolation='nearest')
    dpm_index = data[data['DPM']>a].index
    data['dpm_tag'] = 0
    data.loc[dpm_index, 'dpm_tag']='딜링머신'
    pd.set_option('display.max_rows', None)
    print("==========dpm===============")
    print(data.loc[:,['DPM','dpm_tag']])
    print("============================")
    print(data)
    
    
updateDealingMachine(data)
```

```python
def updateDeanGung(originalData) :
    data = featuring(originalData)
    """
    컬럼별로 랭킹을 구해서 값을 더한 후 상위 5% 추출하기
    GD@15 
    CSD@15
    XPD@15
    Solo Kills
    """
    
    data['GD@15_RANK'] = data["GD@15"].rank(method='max', ascending=False) 
    data['CSD@15_RANK'] = data["CSD@15"].rank(method='max', ascending=False) 
    data['XP@15_RANK'] = data["XPD@15"].rank(method='max', ascending=False) 
    data['SoloKills_RANK'] = data["Solo Kills"].rank(method='max', ascending=False) 
    
    #가중치 어떻게 줄지 설정해야
    data["Daengung Index"] = data['GD@15_RANK']+data['CSD@15_RANK']+data['XP@15_RANK']+data['SoloKills_RANK']+data['Games']*100
    
    
    a = data['Daengung Index'].quantile(q=0.95, interpolation='nearest')
    
    daengGungIndex = data[data['Daengung Index']>a].index 
    data['daengung_tag'] = 0
    
    data.loc[daengGungIndex,'daengung_tag'] = "댕겅파"
    print(data)   
    
    
    """
    댕겅파는 미드로만 할것인가..전체 라인으로 할것인가
    판수 제한을 둬야하나? 멀티포지션이 있군
    """
updateDeanGung(data)
```

```python
def getJungleType(originalData) :
    
    data = featuring(originalData)
    
    
    """
    정글 라인인 사람만 다음 컬럼들의 랭킹을 구해서 값을 더한 후 상위 5% 추출하기
    GD@15 
    CSD@15
    XPD@15
    Solo Kills
    """
    
    mask_isjungle = data['Position'] == "JUNGLE"
    data['JG_GD@15_RANK'] = data[mask_isjungle]['GD@15'].rank(method='max',ascending=False)
    data['JG_CSD@15_RANK'] = data[mask_isjungle]['CSD@15'].rank(method='max',ascending=False)
    data['JG_XP@15_RANK'] = data[mask_isjungle]['XPD@15'].rank(method='max',ascending=False)
    data['JG_SOLO_KILLS'] = data[mask_isjungle]['Solo Kills'].rank(method='max',ascending=False)
        
#   mask_gd = data['GD@15'].rank(method='max',ascending=False) 
#   data[mask_isjungle&mask_gd]
#   data['JG_GD@15_RANK'] = data[isJungle&"GD@15"].rank(method='max', ascending=False) 
#   data['JG_CSD@15_RANK'] = data[isJungle&"CSD@15"].rank(method='max', ascending=False) 
#   data['JG_XP@15_RANK'] = data[isJungle&"XPD@15"].rank(method='max', ascending=False)    
#data['JG_Solo Kills'] = data[isJungle&"Solo Kills"].rank(method='max', ascending=False) 
    
    data["JG_INDEX"] = data['JG_GD@15_RANK']+data['JG_CSD@15_RANK']+data['JG_XP@15_RANK']+data['JG_SOLO_KILLS']
    
    a = data['JG_INDEX'].quantile(q=0.95, interpolation='nearest')
    
    jungleIndex = data[data['JG_INDEX']>a].index 
    data['JG_TAG'] = np.NaN
    data['JG_TAG'][mask_isjungle] = "갱킹형"
    data.loc[jungleIndex,'JG_TAG'] = "성장형"
    
    pd.set_option('display.max_rows', None)
    print(data)   
        
getJungleType(data)
```
